```python
#importação da biblioteca
import pandas as pd
#importação dos dados
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
#normalizando os dados
dados_normalizados = pd.json_normalize(dados['info_moveis'])
dados_normalizados.head()
```

```python
#visualizando as colunas
colunas = list(dados_normalizados.columns)
colunas
```

```python
#retirando as listas
dados_normalizados = dados_normalizados.explode(colunas[3:])
dados_normalizados.head()
```

```python
#retirando o indesx
dados_normalizados = dados_normalizados.reset_index(drop=True)
dados_normalizados.head()
```

```python
#verificando os dados
dados_normalizados.info()
```

```python
#Convertendo os dados numéricos

#importação do numpy
import numpy as np

#trocando o nome do dataframe
dados = dados_normalizados
```

```python
#atribuindo type int a coluna 'max_hospede'
dados['max_hospedes']= dados['max_hospedes'].astype(np.int64)
```

```python
#conferindo o type das colunas
dados.info()
```

```python
#Criando variavel para as colunas
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
#alterando o tipe do objeto das colunas
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
#conferindo
dados.info()
```

```python
#alterando a coluna avaliação geral com o metodo astype
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
#verificando
dados.info()
```

```python
#Tratando da coluna preços
dados['preco']
```

```python
#Substitundo o $ e a ',' por nada
dados['preco'].apply(lambda x: str(x).replace('$', '').replace(',', '').strip())
```

```python
#atribuindo as alterações ao dataframe
dados['preco'] = dados['preco'].apply(lambda x: str(x).replace('$', '').replace(',', '').strip())
```

```python
# alterando o tipo do dado
dados['preco'] =dados['preco'].astype(np.float64)
```
