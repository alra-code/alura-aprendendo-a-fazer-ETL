#**Tratando os dados Vendas clientes**

```python
#importação da biblioteca
import pandas as pd

#leitura do arquivo
dados_vendas_clientes = pd.read_json('/content/dados_vendas_clientes.json')

#visualização do dataframe
dados_vendas_clientes
```

```python
#aplicação do json_normalize
dados_vendas_clientes = pd.json_normalize(dados_vendas_clientes['dados_vendas'])

#visualização do dataframe
dados_vendas_clientes
```

```python
# Coletar os valores das colunas e verificar
colunas = list(dados_vendas_clientes.columns)
colunas
```

```python
# Destrinchar as listas com explode
dados_vendas_clientes = dados_vendas_clientes.explode(colunas[1:])
```

```python
# Resetar os index das linhas
dados_vendas_clientes.reset_index(drop=True,inplace=True)
```

```python
# Observar o DataFrame
dados_vendas_clientes
```

```python
# Verificar os tipos de dados com info
dados_vendas_clientes.info()
```

```python
# A coluna numérica é a 'Valor da compra'
dados_vendas_clientes['Valor da compra']
```

```python
# Remover os textos presentes na base
# Trocar as vírgulas separadoras do decimal por ponto
dados_vendas_clientes['Valor da compra'] = dados_vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$ ', '').replace(',','.').strip())
```

##*Exercício da aula: Tratando números em strings*

```python
import pandas as pd
import numpy as np

# dados da firma
dados = {
    'cargo': ['Gerencia', 'Coordenação', 'Supervisão', 'Analista', 'Assistente', 'Operação', 'Auxiliar', 'Estágio', 'Assessoria', 'Consultoria'],
    'quantidade': ['2 pessoas', '1 pessoa', '3 pessoas', '4 pessoas', '5 pessoas', '3 pessoas', '2 pessoas', '1 pessoa', '1 pessoa', '1 pessoa'],
    'salario': ['$10.000 reais', '$8.000 reais', '$7.000 reais', '$5.000 reais', '$4.000 reais', '$3.500 reais', '$3.000 reais', '$1.500 reais', '$2.500 reais', '$6.000 reais']
}

# transformando o dicionário em DataFrame
df = pd.DataFrame(dados)

# df é o DataFrame com os dados da firma
df
```

```python
df.info()
```

```python
df['quantidade'] = df['quantidade'].apply(lambda x: x.replace(' pessoas', '').replace(' pessoa', ''))
df['salario'] = df['salario'].apply(lambda x: x.replace('$', '').replace(' reais', '').replace('.',''))
```

```python
df['quantidade'] = df['quantidade'].astype(np.int64)
df['salario'] = df['salario'].astype(np.float64)
```
